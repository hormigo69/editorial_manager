```python
import pandas as pd
import openpyxl

# import the sheet 'Manuscritos' from the Excel Editorial Manager.xlsx and store it in a dataframe called editorial_manager_updated
editorial_manager_updated = pd.read_excel('Excel Editorial Manager.xlsx', sheet_name='Manuscritos')

#editorial_manager_updated = pd.read_excel('Excel Editorial Manager.xlsx')

#check the dataframe
editorial_manager_updated
```

```python
# this fucntion calculate some data from the columns obtained from Editorial Manager
import re
df = editorial_manager_updated

#extract the year from the REF column and store it in a new column called año.
# the year are the characters in NEFRO-D-21-00400 between NEFRO-D- and - adding 2000
#use a regular expression to extract the year, convert to int and add 2000
df['año'] = df['REF'].str.extract(r'NEFRO-D-(\d{2})-').astype(int) + 2000

#extract referencia and revision from numero_manuscrito. The REF are the first 14 characters in numero_manuscrito
#use a regular expression to extract the REF and revision
df['referencia'] = df['REF'].str.extract(r'(.{16})')
#revision are the characters in numero_manuscrito after the REF and are in the format Rx where x is a number
df['revision'] = df['REF'].str.extract(r'NEFRO-D-\d{2}-\d{5}-(R\d)')

#remove AÑO
df.drop('AÑO', axis=1, inplace=True)

#in the column Activo, replace the value ⚠️ by True and ✔️ by False
df['Activo'] = df['Activo'].replace({'⚠️': True, '✔️': False})

#in the column > 10 días, add True if Activo is True and Dïas pasado is greater than 10. else add False
df['> 10 días'] = df['Activo'] & (df['Días pasados'] > 10)

#rename all the columns to avois caps
df.columns = df.columns.str.lower()
```

```python
#rename estado actual to estado_actual
df.rename(columns={'fecha _estado': 'fecha_estado'}, inplace=True)
#rename ref maquetad0 to ref_maquetado
df.rename(columns={'ref maquetad0': 'ref_maquetado'}, inplace=True)
# rename revisión to revision
df.rename(columns={'revisión': 'revision_1'}, inplace=True)
#rename coment compl del último revisor to comentarios_ultimo_revisor
df.rename(columns={'coment compl del último revisor': 'comentarios_ultimo_revisor'}, inplace=True)
#rename dec editor para autor to decision_editor
df.rename(columns={'dec editor para autor': 'decision_editor_r1'}, inplace=True)
#rename the column REF to numero_manuscrito
df.rename(columns={'ref': 'numero_manuscrito'}, inplace=True)
#rename aceptación to aceptacion
df.rename(columns={'aceptación': 'aceptacion'}, inplace=True)
#rename maquetación to maquetacion
df.rename(columns={'maquetación': 'maquetacion'}, inplace=True)
#rename in press to in_press
df.rename(columns={'in press': 'in_press'}, inplace=True)
```

```python

import re

#define a function that extract the 'referencia', 'revision' and 'año' from the 'numero_manuscrito'
def extract_ref_rev(x):
    #check if the values in the column are in the correct format: NEFRO-D-##-#####

    # create a regular expression to check if the data is like the pattern "NEFRO-D-##-#####" or "NEFRO-D-##-#####R#"
    pattern = re.compile("^NEFRO-D-[0-9]{2}-[0-9]{5}$|^NEFRO-D-[0-9]{2}-[0-9]{5}R[0-9]$")
    
    #find the values that are not in the correct format and print them
    if not pattern.match(x):
        print('The value is not in the correct format: ' + x)
        ref = 'error'
        rev = 'error'
        year = 'error'

    #extract the reference, revision and year from the column 'numero_manuscrito'
    else:
        #the reference is NEFRO-D-##-#####
        ref = x[0:16]
        #the revision is R# if exist or empty if not
        rev = x[16:19]
        #the year is in the position 9 and 10. Converto to int and add 2000
        year = int(x[8:10]) + 2000

    return ref, rev, year

    
```
